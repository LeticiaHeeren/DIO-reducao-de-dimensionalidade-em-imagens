```python
import numpy as np
import matplotlib.pyplot as plt

def rgb_to_grayscale(image):
    """
    Converte uma imagem colorida (RGB) para níveis de cinza.
    """
    # Fórmula para conversão: gray = 0.2989 * R + 0.5870 * G + 0.1140 * B
    grayscale_image = np.dot(image[..., :3], [0.2989, 0.5870, 0.1140])
    return grayscale_image.astype(np.uint8)  # Garantir que os valores estejam no intervalo 0-255

def binarize_image(grayscale_image, threshold=128):
    """
    Binariza uma imagem em tons de cinza.
    """
    binary_image = np.where(grayscale_image > threshold, 255, 0)
    return binary_image.astype(np.uint8)  # Garantir que os valores sejam 0 ou 255

def load_image(file_path):
    """
    Carrega uma imagem de um arquivo e retorna como um array NumPy.
    """
    from PIL import Image
    image = Image.open(file_path)
    return np.array(image)

def display_images(original, grayscale, binary):
    """
    Exibe as imagens original, em tons de cinza e binarizada.
    """
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.title("Original")
    plt.imshow(original)

    plt.subplot(1, 3, 2)
    plt.title("Tons de Cinza")
    plt.imshow(grayscale, cmap='gray')

    plt.subplot(1, 3, 3)
    plt.title("Binarizada")
    plt.imshow(binary, cmap='gray')

    plt.show()

# Substitua pelo nome exato do arquivo que você fez upload
file_path = "castelo_hohenzollern.jpg"  # Exemplo: use o nome do arquivo que você carregou

# Carregar a imagem
image = load_image(file_path)

# Converter para tons de cinza
grayscale_image = rgb_to_grayscale(image)

# Binarizar a imagem
binary_image = binarize_image(grayscale_image, threshold=128)

# Exibir as imagens
display_images(image, grayscale_image, binary_image)
```
